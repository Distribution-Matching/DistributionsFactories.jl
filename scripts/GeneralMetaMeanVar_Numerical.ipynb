```julia
using Distributions, Plots, Roots
```

```julia
d_target =Weibull(2,3)
```

```julia
x_range = 0:0.01:8
plot(x_range, pdf.(d_target, x_range), label = false)
```

```julia
m, v = mean(d_target), var(d_target)
```

```julia
function dist_from_mean_var(::Type{Gamma}, μ::Number, var::Number)
    α = μ^2/var
    θ = var/μ
    return Gamma(α,θ)
end;
```

```julia
d_approx = dist_from_mean_var(Gamma, m, v)
```

```julia
plot(x_range, pdf.(d_target, x_range), label = "Actual (Weibull)")
plot!(x_range, pdf.(d_approx, x_range), label = "Approximation (Gamma)")
```
